#Autocorrelation

자기 자신을 복제하여 배치한 시그널과 원본 시그널의 correlation.

 = 자신과의 상관관계

사용처: 시그널에서 패턴 찾기, harmonic frequency에서 missing fundamental frequency 찾기 등

패턴을 찾는 것은 복제한 시그널을 시간 차를 두고 비교하면서 찾을 수 있음.

#Root Mean Square

일련의 숫자들의 제곱의 평균에 루트를 씌운 값.

시간에 걸쳐 나타난 값들의 평균을 구하는 데에 유용함.

audio의 경우 amplitude를 제곱하고 시간별 값의 평균을 구한 뒤 그 값의 루트를 구한다.

제곱과 루트를 같이 쓰는 이유는 양수와 음수가 섞여있기 때문이다.
